```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from scipy.stats import linregress
```

### Reading Data from File

```python
df = pd.read_csv('epa-sea-level.csv')
```

```python
df.head()
```

### Create `scatter plot`

```python
fig, ax = plt.subplots()
ax.scatter(df['Year'], df['CSIRO Adjusted Sea Level'])
plt.show()
```

### First line of best fit

```python
lin = linregress(df['Year'], y=df['CSIRO Adjusted Sea Level'])
```

```python
x_fit1 = pd.Series([i for i in range(1880, 2050)])
y_fit1 = lin.slope*x_fit1 + lin.intercept
```

```python
ax.plot(x_fit1, y_fit1, color='red')
fig
```

### Second line of best fit

```python
df_cat = df.loc[df['Year'] >= 2000]
```

```python
lin_2 = linregress(df_cat['Year'], y=df_cat['CSIRO Adjusted Sea Level'])
```

```python
x_fit2 = pd.Series([i for i in range(2000, 2050)])
y_fit2 = lin_2.slope*x_fit2 + lin_2.intercept
```

```python
ax.plot(x_fit2, y_fit2, color='green')
fig
```

### Add labels and title

```python
ax.set_xlabel('Year')
ax.set_ylabel('Sea Level (inches)')
ax.set_title('Rise in Sea Level')
```
